### Q4: Create a new column in the dataframe “age_group”. This column should contain the age group the respondent belongs to. The age groups are 18-25, 25-40, 40-55 and 55+. The dataframe should look like this after the column creation.

```python
survey.head(2)
```

```python
# Make a copy of the dataframe
data = survey.copy()
```

```python
data.head()
```

```python
data.insert(10, 'age_group', survey.age)
```

```python
data.head()
```

```python
data1 = data[(data['age_group'] >= 18) & (data['age_group'] < 25)]
```

```python
data1.head(2)
```

```python
data1['age_group'].unique()
```

```python
data1['age_group'] = '18-25'
```

```python
data1.head(2)
```

```python
data2 = data[(data['age_group'] >= 25) & (data['age_group'] < 40)]
```

```python
data2.head(2)
```

```python
data2['age_group'].unique()
```

```python
data2['age_group'] = '25-40'
```

```python
data2.head(2)
```

```python
data3 = data[(data['age_group'] >= 40) & (data['age_group'] < 55)]
```

```python
data3.head(2)
```

```python
data3['age_group'].unique()
```

```python
data3['age_group'] = '40-55'
```

```python
data3.head(2)
```

```python
data4 = data[data['age_group'] >= 55]
```

```python
data4.head(2)
```

```python
data4['age_group'].unique()
```

```python
data4['age_group'] = '55+'
```

```python
data4.head()
```

```python
# Add all the dataframes
data = pd.concat([data1,data2,data3,data4])
```
